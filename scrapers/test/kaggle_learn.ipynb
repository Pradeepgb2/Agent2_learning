```python
from pathlib import Path
from datetime import datetime
import re
```

```python
BASE_DIR = Path.cwd().resolve().parent # path position of cwd
RAW_DIR = BASE_DIR / 'data' / 'raw'
RAW_DIR= Path(RAW_DIR)
DATASET_PREFIX = "nepal_seismicity_master_"
```

```python
date_pattern = re.compile(r"(\d{4}-\d{2}-\d{2})")
```

```python
files_with_dates = []
```

```python
for file in RAW_DIR.glob(f"{DATASET_PREFIX}*.csv"):
    match = date_pattern.search(file.name)
    if match:
       file_date = datetime.strptime(match.group(1), "%Y-%m-%d")
       files_with_dates.append((file,file_date))
```

```python
files_with_dates.sort(key= lambda x: x[1], reverse=True)
```

```python
print(files_with_dates)
```

```python
if len(files_with_dates)<2:
    new_file, prev_file = files_with_dates[0][0]
else:
    new_file, prev_file = files_with_dates[0][0], files_with_dates[1][0]
    
```
